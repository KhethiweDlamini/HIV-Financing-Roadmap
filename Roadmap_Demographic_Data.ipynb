```python
import pandas as pd
import numpy as np
```

```python
# Read in excel file
file_path = "/content/drive/MyDrive/HIV Roadmap/Further Data/South Africa/Demographic Projecttions - UNDP WPP/Edited_WPP2022_POP_F03_1_POPULATION_SELECT_AGE_GROUPS_BOTH_SEXES.xlsx"
sa_pd = pd.read_excel(file_path, header=1, sheet_name=None)
```

```python
columns_to_keep = ['Variant','Region, subregion, country or area *', 'ISO3 Alpha-code', "Year", "Total", "0-4", "0-19", "65+"]
```

```python
# Dictionary to store filtered and selected data
filtered_sheets = {}
```

```python
# Process each sheet
for sheet_name, sheet_data in sa_pd.items():
    filtered_data = sheet_data[sheet_data['Region, subregion, country or area *'] == "Zambia"][columns_to_keep]
    filtered_sheets[sheet_name] = filtered_data
```

```python
# Save the filtered data to a new Excel file
with pd.ExcelWriter("Zambia_Demographics.xlsx") as writer:
    for sheet_name, filtered_data in filtered_sheets.items():
        filtered_data.to_excel(writer, sheet_name=sheet_name, index=False)
```
